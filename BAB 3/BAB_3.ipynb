**Bab 3: Unsupervised Learning and Preprocessing**

## 1. Tujuan

Bab ini bertujuan untuk memberikan pemahaman yang mendalam tentang metode pembelajaran tak terawasi (unsupervised learning), yang digunakan untuk menganalisis data tanpa memerlukan label atau target yang diketahui. Fokus utama bab ini adalah pada teknik-teknik seperti **clustering**, **reduksi dimensi**, dan **normalisasi data**, yang sangat berguna untuk mengungkap pola atau struktur tersembunyi dalam data. Pembelajaran tak terawasi ini memungkinkan kita untuk menemukan hubungan, kelompok, atau komponen yang dapat membantu dalam analisis lebih lanjut atau dalam membangun model yang lebih kompleks, tanpa memerlukan informasi sebelumnya tentang kategori atau kelas data tersebut.

2. Implementasi Kode
Berikut adalah beberapa contoh kode implementasi untuk metode clustering dan praproses data:

```python
# Mengimpor kelas StandardScaler dari scikit-learn.
# StandardScaler digunakan untuk normalisasi data, yaitu mengubah data sehingga memiliki rata-rata 0 dan standar deviasi 1.
# Teknik ini penting untuk memastikan bahwa fitur-fitur dengan skala yang berbeda tidak mendominasi model pembelajaran mesin.
from sklearn.preprocessing import StandardScaler

# Mengimpor fungsi load_iris untuk memuat dataset Iris dari scikit-learn.
# Dataset Iris berisi data mengenai spesies bunga dengan fitur-fitur seperti panjang dan lebar kelopak serta mahkota.
from sklearn.datasets import load_iris
```

```python
# Memuat dataset Iris menggunakan fungsi load_iris() dari scikit-learn.
# Dataset Iris adalah dataset klasik yang sering digunakan untuk eksperimen klasifikasi.
# Data ini terdiri dari fitur seperti panjang dan lebar kelopak serta mahkota bunga.
iris = load_iris()

# Mengambil data fitur (X) dari dataset Iris.
# Iris['data'] berisi array 2D dengan ukuran (150, 4), yang berisi 150 sampel dan 4 fitur untuk masing-masing sampel.
data = iris['data']
```

```python
# Membuat objek StandardScaler untuk melakukan normalisasi data.
# StandardScaler akan mengubah setiap fitur sehingga memiliki rata-rata 0 dan standar deviasi 1.
scaler = StandardScaler()

# Melakukan normalisasi (scaling) pada data fitur (data) menggunakan metode fit_transform().
# fit_transform() akan menghitung rata-rata dan standar deviasi dari data, kemudian mengaplikasikan normalisasi pada data tersebut.
data_scaled = scaler.fit_transform(data)

# Menampilkan 5 baris pertama dari data sebelum scaling untuk membandingkan.
print("Data sebelum scaling:\n", data[:5])

# Menampilkan 5 baris pertama dari data setelah scaling untuk melihat perubahan pada skala fitur.
print("Data setelah scaling:\n", data_scaled[:5])
```

```python
# Mengimpor kelas KMeans dari scikit-learn.
# KMeans adalah algoritma clustering yang digunakan untuk membagi data ke dalam kelompok (cluster) berdasarkan kedekatannya.
# Algoritma ini mencoba meminimalkan jarak antara data dengan pusat cluster yang ditentukan oleh nilai k.
from sklearn.cluster import KMeans
```

```python
# Membuat objek KMeans dengan parameter n_clusters=3.
# n_clusters=3 menunjukkan bahwa kita ingin membagi data ke dalam 3 kelompok (cluster).
# random_state=0 digunakan untuk memastikan hasil yang konsisten setiap kali kode dijalankan, sehingga hasil clustering dapat direproduksi.
kmeans = KMeans(n_clusters=3, random_state=0)

# Melatih model KMeans pada data yang sudah dinormalisasi (data_scaled).
# Algoritma k-Means akan mengelompokkan data ke dalam 3 cluster berdasarkan kedekatannya dengan pusat cluster.
kmeans.fit(data_scaled)
```

```python
# Menampilkan koordinat pusat dari setiap cluster yang ditemukan oleh k-Means.
# cluster_centers_ berisi nilai rata-rata dari fitur untuk setiap cluster yang terbentuk.
# Setiap baris mewakili pusat cluster yang sesuai dengan urutan cluster yang dihasilkan.
print("Cluster centers:\n", kmeans.cluster_centers_)

# Menampilkan label untuk setiap data point.
# labels_ adalah array yang menunjukkan cluster mana yang menjadi tempat setiap sampel data.
# Setiap elemen dalam labels_ berisi angka yang menunjukkan nomor cluster yang dipilih untuk sampel data tersebut.
print("Labels untuk setiap data:\n", kmeans.labels_)
```

```python
# Mengimpor kelas PCA (Principal Component Analysis) dari scikit-learn.
# PCA adalah teknik reduksi dimensi yang digunakan untuk mengurangi jumlah fitur dalam dataset.
# PCA melakukan transformasi data ke ruang dimensi yang lebih rendah, sambil mempertahankan sebanyak mungkin variasi dalam data.
# Teknik ini sangat berguna ketika bekerja dengan data berdimensi tinggi dan ingin mengurangi kompleksitas atau visualisasi.
from sklearn.decomposition import PCA
```

```python
# Membuat objek PCA dengan n_components=2.
# n_components=2 berarti kita ingin mereduksi dimensi data menjadi 2 komponen utama.
# PCA akan mencari dua dimensi baru yang paling banyak menjelaskan variansi dalam data.
pca = PCA(n_components=2)

# Melakukan reduksi dimensi pada data yang sudah dinormalisasi (data_scaled) dengan metode fit_transform().
# fit_transform() akan menghitung dua komponen utama dan mentransformasikan data ke dalam ruang dimensi dua.
data_pca = pca.fit_transform(data_scaled)

# Menampilkan 5 baris pertama dari data setelah reduksi dimensi.
# Data kini hanya memiliki dua fitur (komponen utama) yang menjelaskan variasi terbesar dalam data.
print("Data setelah reduksi dimensi:\n", data_pca[:5])
```

```python
# Mengimpor pustaka matplotlib.pyplot untuk visualisasi data.
# Matplotlib adalah pustaka untuk membuat grafik dan plot di Python.
# pyplot berfungsi untuk membuat plot sederhana seperti scatter plot, bar chart, dll.
import matplotlib.pyplot as plt
```

```python
# Mendefinisikan fungsi untuk membuat visualisasi scatter plot berdasarkan data dan label.
# Fungsi ini akan memplot data dalam dua dimensi (komponen utama yang telah direduksi).
# Warna titik pada plot akan menggambarkan cluster yang diberikan oleh k-Means.
def plot_clusters(data, labels, title):
    # Membuat scatter plot dengan data pada komponen utama 1 dan 2.
    # c=labels akan mewarnai titik berdasarkan label cluster yang diberikan oleh k-Means.
    # cmap='viridis' menentukan skema warna yang digunakan untuk representasi cluster.
    # s=50 mengatur ukuran titik pada plot.
    plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', s=50)

    # Menambahkan judul pada plot.
    plt.title(title)

    # Menambahkan label untuk sumbu X dan Y.
    plt.xlabel('Komponen Utama 1')
    plt.ylabel('Komponen Utama 2')

    # Menambahkan colorbar untuk melihat representasi warna yang digunakan untuk label.
    plt.colorbar()

    # Menampilkan plot.
    plt.show()

# Memanggil fungsi plot_clusters untuk visualisasi data yang telah direduksi dimensinya,
# dengan label cluster yang diperoleh dari hasil clustering k-Means.
plot_clusters(data_pca, kmeans.labels_, "Hasil Clustering k-Means")
```

## 3. Penjelasan Teoritis

### 3.1 Apa itu Pembelajaran Tak Terawasi?
Pembelajaran tak terawasi (unsupervised learning) adalah metode pembelajaran mesin di mana model dilatih menggunakan data yang tidak memiliki label atau target yang diketahui. Tujuannya adalah untuk mengidentifikasi pola atau struktur yang tersembunyi dalam data, seperti pengelompokan atau pengurangan dimensi. Karena tidak ada informasi target, algoritma ini berfokus pada menemukan kesamaan atau perbedaan antar data.

### 3.2 Teknik Praproses Data
- **Normalisasi**: Proses ini penting agar semua fitur dalam dataset memiliki skala yang serupa. Normalisasi memastikan bahwa setiap fitur memiliki distribusi dengan rata-rata 0 dan varians 1. Hal ini membantu mencegah fitur dengan rentang nilai yang lebih besar mendominasi model, terutama dalam algoritma berbasis jarak seperti k-Means.
- **Reduksi Dimensi**: Teknik ini digunakan untuk mengurangi jumlah fitur dalam data, sehingga lebih mudah dianalisis atau divisualisasikan. Dengan mengurangi dimensi, kita juga dapat mengurangi kompleksitas komputasi, serta mengurangi kemungkinan terjadinya overfitting.

### 3.3 Metode Clustering
- **k-Means**: Merupakan metode clustering yang membagi data ke dalam k kelompok berdasarkan kedekatannya dengan pusat cluster. Setiap data point akan dimasukkan ke dalam cluster yang memiliki pusat terdekat. Proses ini diulang hingga posisi pusat cluster stabil.
- **Evaluasi Clustering**: Untuk menilai seberapa baik clustering dilakukan, kita bisa menggunakan metrik seperti **Silhouette Score**, yang mengukur seberapa mirip setiap data point dengan data point di cluster yang sama dibandingkan dengan cluster lain.

### 3.4 PCA untuk Reduksi Dimensi
Principal Component Analysis (PCA) adalah teknik statistik yang digunakan untuk mengurangi dimensi data dengan cara mencari komponen utama yang menjelaskan variasi terbesar dalam data. PCA mengubah data asli ke dalam ruang fitur yang baru, dengan komponen utama sebagai sumbu baru. Hal ini memungkinkan kita untuk memvisualisasikan data dalam dimensi yang lebih rendah (misalnya, 2 dimensi) tanpa kehilangan terlalu banyak informasi.

- **PCA**: Reduksi dimensi dengan PCA berhasil mengubah data berdimensi tinggi menjadi 2 dimensi yang mudah dianalisis dan divisualisasikan. Hal ini menunjukkan bahwa PCA efektif dalam mempertahankan variasi data penting sambil menyederhanakan kompleksitas.

Dengan menggunakan metode ini, kita dapat mengungkap struktur tersembunyi dalam data dan menyederhanakan analisis data yang lebih kompleks.
